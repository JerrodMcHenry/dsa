# Classes and Pointers


## Classes

The constuctor is the first part of the class Cookie. The self keyword means it is a method that is part of a class. Color is a parameter and gets passed to the specific instance of color we are creating for the cookie. 

```python
# This is a constructor - 

class Cookie:
    def __init__(self, color):
        self.color = color
```

```python
class Cookie:
    def __init__(self, color): #Consructor 
        self.color = color

    def get_color(self): # 1st method
        return self.color
    
    def set_color(self,color): # 2nd method 
        self.color = color

cookie_one = Cookie("green")
cookie_two = Cookie("blue")

print("Cookie one is", cookie_one.get_color())
print("Cookie two is", cookie_two.get_color())

cookie_one.set_color("yellow")

print("\nCookie one is now", cookie_one.get_color())
print("Cookie two is still", cookie_two.get_color())
```

# Pointers 

When we assign num2 = num1 they are both pointing at the same address in memory. They do not create a new place in memory 

```python
num1 = 11
num2 = num1

print("Before num2 value is updated:")
print("num1 =", num1)
print("num2 =", num2)

print("\nnum1 points to:", id(num1))
print("num2 points to:", id(num2))
```

# Integers are immutable therefore they cannot be changed. 

Once number 11 is put in a particular place in memory you cannot change that number 11. You can point num1 to a different integer that is 
stored somewhere else but you cant actually change the number 11 once you create it. 

```python
num1 = 11
num2 = num1

print("Before num2 value is updated:")
print("num1 =", num1)
print("num2 =", num2)

print("\nnum1 points to:", id(num1))
print("num2 points to:", id(num2))

num2 = 22

print("\nAfter num2 value is updated:")
print("num1 = ", num1)
print("num2 = ", num2)

print("\nnum1 points to:", id(num1))
print("num2 points to:", id(num2))
```

# Dictionaries are mutable and CAN be changed. 

3 dictionaies can all be changed and point to the same address. Python will automatically us "Garbage Collection" to remove old dictionary addresses that are
no longer being pointed to. 

```python
dict1 = {
    "value": 11
}

dict2 = dict1

print("Before value is updated:")
print("dict1 =", dict1)
print("dict2 =", dict2)

print("\ndict1 points to:", id(dict1))
print("dict2 points to:", id(dict2))

dict2['value'] = 22

print("\nAfter value is updated:")
print("dict1 =", dict1)
print("dict2 =", dict2)

print("\ndict1 points to:", id(dict1))
print("dict2 points to:", id(dict2))
```
