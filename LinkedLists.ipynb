# Linked List Intro

```python
Linked Lists are diffent than regular lists. Linked lists do NOT have indexes like a regular list. The nodes in a linked list are spread out in memory. A variable called "head" points to the first node in the linked list and tail
points to the last node in the linked list. Each node within the list points to the next and the last one points to none. 
```

# Appending a new node to the end of the list

The last node points directly to the new node at the end and so does tail. This happens in 1 operation therefore it is O(1)


![Screenshot 2026-01-04 at 8.54.06 PM.png](<attachment:Screenshot 2026-01-04 at 8.54.06 PM.png>)

# Removing a node from the end of the list 

When we remove from the end of the linked list it is complicated. In order to have tail point to the last node, the tail pointer has to be set equal to another pointer that is pointing at the same node. So in order to get to that node we have to start at the head and iterate through the list until we get to that pointer. Then we can set tail equal to that pointer. 

Because we had to iterate through the entire list it is O(n)
