```python
Linked Lists are diffent than regular lists. Linked lists do NOT have indexes like a regular list. The nodes in a linked list are spread out in memory. A variable called "head" points to the first node in the linked list and tail
points to the last node in the linked list. Each node within the list points to the next and the last one points to none. 
```
