# What is Big O?

Big O is a way to compare 2 sets of code. If code 1 and code 2 both acomplish the same thing. We use Big O to compare both codes mathmatically to determine how efficient they run

# Time Complexity 

If code 1 took 15 seconds to complete but code 2 took 1 minute to complete then you can measure the effeiency by using time complexity. 

```python
Time complexity is NOT measured in time. If you took the code and ran it on a computer that is twice as fast as the other computer than it would complete twice as fast. 
So it wouldnt make the code any better, it just means the computer is better.

Time Complexity is measured by the number of operations that it takes to complete something. 
```
