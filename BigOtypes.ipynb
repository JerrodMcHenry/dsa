# O(n)

```python
O of n is always going to be a straight line. It is proportional. Its called the linear approach because the runtime grows directly proportional to the input size
```

```python
def print_items(n):
    for i in range(n):
        print(i)

print_items(10)
```

# O(n^2)

O of n^2 line is a lot steeper than o(n). This means its a lot less efficient from a time complexity standpoint. This is called the quadratic approach because the runtime grows like the square of the input size

```python
def print_items(n):
    for i in range(n):
        for j in range(n):
            print(i,j)


print_items(10)
```
